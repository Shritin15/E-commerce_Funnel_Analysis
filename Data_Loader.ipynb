```python
import pandas as pd
from sqlalchemy import create_engine

connection_string = 'postgresql://postgres:Shritin15@localhost:5432/Funnel_Analysis'
engine = create_engine(connection_string)

files = ['2019-Oct.csv', '2019-Nov.csv']

for file_path in files:
    for i, chunk in enumerate(pd.read_csv(file_path, chunksize=50000)):
        
        chunk.columns = chunk.columns.str.lower().str.replace('.', '_')
        
       
        chunk['event_time'] = pd.to_datetime(chunk['event_time']).dt.tz_localize(None)
        
        
        if i == 0 and file_path == '2019-Oct.csv':
            chunk.to_sql('user_behavior', engine, if_exists='replace', index=False)
        else:
            chunk.to_sql('user_behavior', engine, if_exists='append', index=False)
            
        print(f"Processed chunk {i+1} from {file_path}")
```
